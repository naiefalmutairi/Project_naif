***مشروع التنبؤ بدرجات الطالب***

```python
# استيراد المكتبات
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# البيانات
data = {
    "Hours": [1, 2, 3, 4, 5, 6],
    "Scores": [40, 50, 65, 70, 80, 90]
}
df = pd.DataFrame(data)

X = df[["Hours"]]
y = df["Scores"]
```

```python
# بناء النموذج
model = LinearRegression()
model.fit(X, y)

# التنبؤ
predicted_score = model.predict([[7]])
print("درجة التوقع لطالب ذاكر 7 ساعات:", predicted_score[0])
```
