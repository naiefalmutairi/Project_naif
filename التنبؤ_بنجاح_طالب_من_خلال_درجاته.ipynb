***التنبؤ بنجاح طالب من خلال درجاته***

تطبيق عملي: التنبؤ بنجاح طالب من خلال درجاته
(باستخدام الانحدار الخطي Linear Regression)

الهدف اليوم: بناء نموذج تعلم آلي بسيط باستخدام مكتبة Scikit-learn يتنبأ بدرجات الطالب النهائية اعتمادًا على ساعات الدراسة.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# مكتبات دعم اللغة العربية
import arabic_reshaper
from bidi.algorithm import get_display

# دالة لإصلاح اتجاه النص العربي
def arabic_text(text):
    reshaped = arabic_reshaper.reshape(text)
    return get_display(reshaped)
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'DejaVu Sans'

# بيانات تجريبية
data = {
    'ساعات المذاكرة': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'النتيجة النهائية': [52, 55, 61, 70, 73, 78, 81, 90, 95]
}
df = pd.DataFrame(data)

# نموذج الانحدار الخطي
X = df[['ساعات المذاكرة']]
y = df['النتيجة النهائية']
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)

# الرسم البياني
plt.figure(figsize=(8, 5))
plt.scatter(df['ساعات المذاكرة'], df['النتيجة النهائية'], color='blue')
plt.plot(df['ساعات المذاكرة'], predictions, color='red')

# عناوين ومحاور بالعربية
plt.title(arabic_text('العلاقة بين ساعات المذاكرة والنتيجة النهائية'))
plt.xlabel(arabic_text('ساعات المذاكرة'))
plt.ylabel(arabic_text('النتيجة النهائية'))

plt.grid(True)
plt.tight_layout()
plt.show()
```
